```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt 
import cv2
import numpy as np 
```

## Brightness enhancement

**Brightness enhancement is a technique used to adjust the brightness level of an image. It's commonly used to make the image brighter or darker. In brightness enhancement, the pixel values of an image are adjusted to increase or decrease the overall brightness of the image. This can be done by multiplying each pixel value by a constant factor or by adding a constant value to each pixel**

```python
im = Image.open('pollen_grains.tif')

curr_bri = ImageEnhance.Brightness(im)
new_bri = 2.1
  
img_brightened = curr_bri.enhance(new_bri)

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(im,cmap='gray')
ax1.set_title("  Original Pollen")
ax2.imshow(img_brightened,cmap='gray')
ax2.set_title(" Enhanced Pollen")

plt.show();
```

## Brightness slicing

**It is a technique used to highlight specific features of an image by isolating a range of pixel intensities. It's commonly used in medical imaging to identify regions of interest, such as tumors or blood vessels. In intensity slicing, a range of pixel intensities is selected, and all pixels within that range are set to a maximum value, while all other pixels are set to a minimum value. This creates a binary image that highlights the regions of interest in the original image**

```python
def Brightnessslicing(image,Imin,Imax):
    image = cv2.GaussianBlur(image, (5,5),0)
    width,height = image.shape
    x = np.zeros((width,height),dtype = 'int8')
    for i in range(width):
        for j in range(height):
            if (image[i,j]>Imin and image[i,j]<Imax):
                x[i,j] = 255
                print(x[i,j])
            else:
                x[i,j] = image[i,j]

    fig, (ax1, ax2) = plt.subplots(1, 2)

    ax1.imshow(image,cmap = 'gray',vmin = 0,vmax = 255)
    ax1.set_title("  Original Image")
    ax2.imshow(x,cmap = 'gray',vmin = 0,vmax = 255)
    ax2.set_title('Brighter Sliced Image')
```

```python
image = cv2.imread('kidney.tif')
image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
Brightnessslicing(image,100,100)
```
